```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://quotes.toscrape.com/page/1/" 
response = requests.get(url) 
response = response.content 
soup = BeautifulSoup(response,'html.parser') 
contents = soup.find_all('div',class_ = 'quote')
```

```python
for cont in contents:
    quotes = cont.find('span',class_ = 'text').text
    quotes = quotes[1:-2]
    print(quotes)
```

```python
for cont in contents:
    name = cont.find('small',class_ = 'author').text
    name = name[0:]
    print(name)
```

```python
for cont in contents:
    tags = cont.find('meta')
    tags = tags.attrs['content']
    print(tags)
```

```python
contents = soup.find_all('div',class_ = 'quote')
result = []
for cont in contents:
    quotes = cont.find('span',class_ = 'text').text
    quotes = str(quotes[1:-2])
    name = cont.find('small',class_ = 'author').text
    name = str(name[0:])
    tags = cont.find('meta')
    tags = tags.attrs['content']
    result.append([quotes,name,tags])
```

```python
result = []
for i in range(1,5):
    url = "http://quotes.toscrape.com/page/1/" 
    response = requests.get(url) 
    response = response.content 
    soup = BeautifulSoup(response,'html.parser')
    contents = soup.find_all('div',class_ = 'quote')
    for cont in contents:
        quotes = cont.find('span',class_ = 'text').text
        quotes = str(quotes[1:-2])
        name = cont.find('small',class_ = 'author').text
        name = str(name[0:])
        tags = cont.find('meta')
        tags = tags.attrs['content']
        result.append([quotes,name,tags])
```

```python
result
```

```python
df = pd.DataFrame(result,columns = ['NAME','QUOTES','TAGS'])
```
